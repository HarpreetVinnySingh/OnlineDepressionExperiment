```python
from pycaret.datasets import get_data
from pycaret.classification import *
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Machine_Learning/Training_Data.csv')

exp = setup(
    data,
    target = "Health_group",
    normalize = True
)
best_model = compare_models(fold = 3, sort = 'AUC')
```

```python

tuned = tune_model(best_model)
```

```python

df2 = pd.read_csv('/content/drive/MyDrive/Machine_Learning/Evaluation_Data.csv')
df2.dropna()


predict_model(tuned, df2)
```

```python
finalize_model(tuned)
```

```python
save_model(tuned, model_name='Extra-Trees-Classifier_HS')
```

```python
evaluate_model(tuned)
```
